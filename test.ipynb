```python
import pandas as pd
import numpy as np
df = pd.read_csv('laptopData.csv')
df.head()

# remove the unnamed column

df = df.drop('Unnamed: 0', axis=1)
df.head()
# check for the null values and remove them
df.isna().sum()
df.dropna(inplace=True)
# clean the Ram column
df['Ram'] = df['Ram'].astype(str).str.replace('GB', '').str.strip().astype(int)
# clean the weight string
df['Weight'] = (
    df['Weight']
    .astype(str)
    .str.replace('kg', '', regex=False)
    .str.replace('?', '', regex=False)
    .str.strip()
)

df['Weight'] = pd.to_numeric(df['Weight'], errors='coerce')

df.dtypes
# Filtering
df[(df['TypeName'] == 'Gaming') | (df['Price'] < 60000)]

subset_df = df[df['Company'].isin(['Apple', 'HP', 'Dell'])]
subset_df = subset_df.sort_values(by='Price', ascending=False)
subset_df
# average price of each company laptop

df.groupby('Company')['Price'].mean().sort_values(ascending=False)

df.groupby('Company').agg({
    "Price": ["min", "max", "mean"]
})
df['Inches'] = df['Inches'].astype(str).replace("?", None).astype(float)
df['Inches'] = pd.to_numeric(df['Inches'], errors='coerce')
df['Price_per_inch'] = df['Price'] / df['Inches']

def check_func(screen):
    if "Touchscreen" in screen:
        return "Touch"
    else:
        return "Standard"

df['ScreenType'] = df['ScreenResolution'].apply(check_func)
df
```
